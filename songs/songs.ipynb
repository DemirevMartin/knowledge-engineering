### 1.5 Normalization Function

A shared text normalization applied to both Grammy and WASABI fields before matching:
1. Lowercase
2. Strip parenthetical suffixes like `(Album)`, `(feat. X)`, `(Deluxe Edition)`, `(Remastered)`
3. Remove all punctuation
4. Collapse whitespace

```python
_PAREN_RE = re.compile(r"\s*[\(\[].+?[\)\]]")   # strip (...) and [...]
_PUNCT_RE  = re.compile(r"[^\w\s]")              # remove non-word non-space
_SPACE_RE  = re.compile(r"\s+")                  # collapse whitespace

def normalize(text: str) -> str:
    """Lowercase → strip parentheticals → remove punctuation → collapse spaces."""
    if not isinstance(text, str):
        return ""
    t = text.lower()
    t = _PAREN_RE.sub("", t)
    t = _PUNCT_RE.sub("", t)
    t = _SPACE_RE.sub(" ", t).strip()
    return t

# ── Verify the function ────────────────────────────────────────────────────────
test_cases = [
    "Fearless (Taylor's Version)",
    "Blinding Lights (feat. The Weeknd)",
    "Abbey Road (Album)",
    "What's Going On",
    "  The  Beatles  ",
]
for tc in test_cases:
    print(f"  {tc!r:45s} → {normalize(tc)!r}")
```

## Phase 2 — Matching Grammy Nominations to WASABI

**Strategy (exact-match on normalized text):**

For each Grammy work-type row we have a `(artist_norm, work_title_norm)` pair.  
We try two passes against WASABI:

1. **Song match** — `(artist_norm == wasabi.artist_norm) AND (work_title_norm == wasabi.title_norm)`
2. **Album match** — `(artist_norm == wasabi.artist_norm) AND (work_title_norm == wasabi.album_norm)`

A Grammy row is considered *matched* if either pass succeeds.  
When multiple WASABI rows satisfy the condition (e.g., the same artist has the same album title twice) we keep the first WASABI record returned.

No fuzzy matching is used — this keeps false-positive rate low at the cost of coverage.

### 2.1 Build WASABI Lookup Indices

Build two dictionaries keyed by `(artist_norm, title_norm)` and `(artist_norm, album_norm)`.  
This converts O(N×M) row-by-row scanning into O(1) hash lookups.

```python
# Columns we want to bring over from WASABI into the matched dataset
WASABI_KEEP = [
    "title", "artist", "albumTitle",
    "genre", "album_genre",
    "bpm", "gain",
    "publicationDate",
    "language",
    # normalized keys (useful for diagnostics)
    "title_norm", "album_norm", "artist_norm",
]

# Subset to only the columns we need (saves memory during index builds)
ws = wasabi_songs[WASABI_KEEP].copy()

# ── Index 1: (artist_norm, title_norm) → first matching row index ─────────────
song_idx = {}
for row_idx, (art, ttl) in enumerate(zip(ws["artist_norm"], ws["title_norm"])):
    key = (art, ttl)
    if key not in song_idx:
        song_idx[key] = row_idx

# ── Index 2: (artist_norm, album_norm) → first matching row index ─────────────
album_idx = {}
for row_idx, (art, alb) in enumerate(zip(ws["artist_norm"], ws["album_norm"])):
    key = (art, alb)
    if key not in album_idx:
        album_idx[key] = row_idx

print(f"Song-title index entries : {len(song_idx):,}")
print(f"Album-title index entries: {len(album_idx):,}")
```

### 2.2 Run Matching

For each Grammy row: try song-title lookup first, then fall back to album-title lookup.  
Record which pass succeeded (`match_type`: `"song"` or `"album"`).  
Rows that match neither are left unmatched.

```python
ws_records = ws.to_dict("records")  # fast row access by index

matched_rows = []

for _, row in grammys.iterrows():
    g_art  = row["artist_norm"]
    g_work = row["work_title_norm"]

    wasabi_row = None
    match_type = None

    # Pass 1: song-title match
    idx = song_idx.get((g_art, g_work))
    if idx is not None:
        wasabi_row = ws_records[idx]
        match_type = "song"
    else:
        # Pass 2: album-title match
        idx = album_idx.get((g_art, g_work))
        if idx is not None:
            wasabi_row = ws_records[idx]
            match_type = "album"

    if wasabi_row:
        matched_rows.append({
            # Grammy fields
            "grammy_year"    : row["year"],
            "category"       : row["category"],
            "work_title"     : row["work_title"],
            "artist"         : row["artist"],
            "winner"         : row["winner"],
            # Match metadata
            "match_type"     : match_type,
            # WASABI fields
            "wasabi_title"   : wasabi_row["title"],
            "wasabi_artist"  : wasabi_row["artist"],
            "wasabi_album"   : wasabi_row["albumTitle"],
            "genre"          : wasabi_row["genre"],
            "album_genre"    : wasabi_row["album_genre"],
            "bpm"            : wasabi_row["bpm"],
            "gain"           : wasabi_row["gain"],
            "publication_date": wasabi_row["publicationDate"],
            "language"       : wasabi_row["language"],
        })

matched_df = pd.DataFrame(matched_rows)
print(f"Grammy work-type rows : {len(grammys):,}")
print(f"Matched rows          : {len(matched_df):,}")
print(f"Unmatched rows        : {len(grammys) - len(matched_df):,}")
print(f"Overall match rate    : {len(matched_df)/len(grammys)*100:.1f}%")
print(f"\nMatch type breakdown:")
print(matched_df["match_type"].value_counts())
```

### 3.2 Category Stratification

Map the 764 Grammy category names to broad genre groups using keyword matching, then report which groups are **robust** (≥20 matched records) and which are **sparse** (<20 matched records).

```python
import re

# Keyword → genre group mapping (ordered: first match wins)
CATEGORY_GROUPS = [
    ("Hip-Hop/Rap",   r"rap|hip.?hop|urban"),
    ("R&B/Soul",      r"\br&b\b|rhythm.{0,5}blues|soul|funk|neo.?soul"),
    ("Country",       r"country|bluegrass|americana|folk"),
    ("Rock",          r"\brock\b|metal|punk|alternative|grunge|indie"),
    ("Pop",           r"\bpop\b"),
    ("Electronic",    r"electronic|dance|techno|edm|house|trance|ambient"),
    ("Jazz",          r"\bjazz\b|bebop|fusion|big.?band|swing"),
    ("Classical",     r"classical|orchestr|symphony|chamber|opera|concerto|quartet|ensemble"),
    ("Gospel/CCM",    r"gospel|christian|inspirational|spiritual|contemporary.{0,10}christian|ccm"),
    ("Latin",         r"\blatin\b|salsa|bossa|samba|cumbia|tango|reggaeton"),
    ("Reggae",        r"reggae|ska|dancehall"),
    ("Blues",         r"\bblues\b"),
    ("Children/Spoken", r"children|spoken|comedy|humor|musical theater|theatrical"),
    ("New Age",       r"new.?age"),
    ("World",         r"world|afro|african|celtic|global"),
]

def assign_group(category: str) -> str:
    cat_lower = category.lower()
    for group, pattern in CATEGORY_GROUPS:
        if re.search(pattern, cat_lower):
            return group
    return "Other/General"

# Apply to both Grammy full set and matched set
grammys["category_group"]   = grammys["category"].map(assign_group)
matched_df["category_group"] = matched_df["category"].map(assign_group)

print("Category group distribution in full Grammy set:")
print(grammys["category_group"].value_counts().to_string())
print(f"\nTotal Grammy work-type rows: {len(grammys):,}")
```

```python
SPARSE_THRESHOLD = 20  # fewer than this → flagged as sparse

# Per category group: total Grammy nominations, matched count, match rate, robustness flag
strat = (
    grammys.groupby("category_group").size().rename("grammy_total")
    .to_frame()
    .join(matched_df.groupby("category_group").size().rename("matched"), how="left")
    .fillna(0)
    .astype({"matched": int})
)
strat["match_%"]  = (strat["matched"] / strat["grammy_total"] * 100).round(1)
strat["status"]   = strat["matched"].apply(
    lambda n: "✅ Robust" if n >= SPARSE_THRESHOLD else "⚠️  Sparse"
)
strat = strat.sort_values("matched", ascending=False)

print("Category group stratification (match coverage):")
print(f"{'Group':<22} {'Grammy':>7} {'Matched':>8} {'Match%':>7}  Status")
print("-" * 60)
for grp, row in strat.iterrows():
    print(f"{grp:<22} {row['grammy_total']:>7,} {row['matched']:>8,} {row['match_%']:>6.1f}%  {row['status']}")

robust_groups = strat[strat["status"] == "✅ Robust"].index.tolist()
sparse_groups = strat[strat["status"] == "⚠️  Sparse"].index.tolist()

print(f"\nRobust groups ({len(robust_groups)}): {robust_groups}")
print(f"Sparse groups ({len(sparse_groups)}): {sparse_groups}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

strat_plot = strat.sort_values("matched")
colors = ["#d73027" if s == "⚠️  Sparse" else "#4393c3" for s in strat_plot["status"]]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: matched count per category group
axes[0].barh(strat_plot.index, strat_plot["matched"], color=colors)
axes[0].set_xlabel("Matched records")
axes[0].set_title("Matched Grammy Records by Category Group")
for i, (n, s) in enumerate(zip(strat_plot["matched"], strat_plot["status"])):
    axes[0].text(n + 5, i, str(n), va="center", fontsize=8)
robust_patch = mpatches.Patch(color="#4393c3", label=f"Robust (≥{SPARSE_THRESHOLD})")
sparse_patch  = mpatches.Patch(color="#d73027", label=f"Sparse (<{SPARSE_THRESHOLD})")
axes[0].legend(handles=[robust_patch, sparse_patch])

# Right: match rate (%)
rate_colors = ["#d73027" if s == "⚠️  Sparse" else "#4dac26" for s in strat_plot["status"]]
axes[1].barh(strat_plot.index, strat_plot["match_%"], color=rate_colors)
axes[1].set_xlabel("Match rate (%)")
axes[1].set_title("Grammy → WASABI Match Rate by Category Group")
axes[1].axvline(25.1, color="black", linestyle="--", linewidth=0.8, label="Overall 25.1%")
for i, v in enumerate(strat_plot["match_%"]):
    axes[1].text(v + 0.3, i, f"{v:.0f}%", va="center", fontsize=8)
axes[1].legend()

plt.suptitle("Phase 3: Category Stratification", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
import requests
import time
import json
from pathlib import Path

# ── ⚠️  FILL IN YOUR API KEY HERE ─────────────────────────────────────────────
LASTFM_API_KEY = "36f68c5c91e42ffa588b63bcf2379c55"
# ──────────────────────────────────────────────────────────────────────────────

LASTFM_BASE    = "https://ws.audioscrobbler.com/2.0/"
CACHE_PATH     = Path(DATA_DIR) / "lastfm_cache.json"
RATE_LIMIT_SEC = 0.22   # ~4.5 req/sec (safely under 5/sec limit)

# Load existing cache (allows safe resume after interruption)
if CACHE_PATH.exists():
    with open(CACHE_PATH, "r", encoding="utf-8") as fh:
        _cache = json.load(fh)
    print(f"Loaded {len(_cache):,} cached entries from {CACHE_PATH.name}")
else:
    _cache = {}
    print("Starting fresh cache.")

def _cache_key(artist: str, title: str) -> str:
    return f"{artist.lower().strip()}|||{title.lower().strip()}"

def query_lastfm(artist: str, title: str) -> dict:
    """Query Last.fm track.getInfo. Returns dict with 'listeners' and 'playcount' (or None on miss)."""
    key = _cache_key(artist, title)
    if key in _cache:
        return _cache[key]

    params = {
        "method"  : "track.getInfo",
        "api_key" : LASTFM_API_KEY,
        "artist"  : artist,
        "track"   : title,
        "format"  : "json",
        "autocorrect": "1",
    }
    try:
        resp = requests.get(LASTFM_BASE, params=params, timeout=10)
        resp.raise_for_status()
        data = resp.json()
        track = data.get("track", {})
        result = {
            "listeners": int(track.get("listeners", 0)) if track.get("listeners") else None,
            "playcount": int(track.get("playcount", 0)) if track.get("playcount") else None,
        }
    except Exception:
        result = {"listeners": None, "playcount": None}

    _cache[key] = result
    time.sleep(RATE_LIMIT_SEC)
    return result

if LASTFM_API_KEY == "YOUR_API_KEY_HERE":
    print("⚠️  API key not set. Fill in LASTFM_API_KEY above, then re-run this cell before proceeding.")
else:
    print(f"✅ API key configured. Ready to query.")
```

### 3.4 Run Last.fm Enrichment

Queries unique (artist, title) pairs only — deduplicating saves ~30% of API calls.  
Saves cache every 100 queries. **Estimated time: ~25–35 min** (~4,300 unique pairs at 4.5 req/sec).  
Re-running this cell after an interruption will resume from the cache automatically.

```python
from tqdm.auto import tqdm

assert LASTFM_API_KEY != "YOUR_API_KEY_HERE", \
    "Set your API key in the cell above first, then re-run that cell, then run this one."

# Unique (artist, title) pairs — query each once regardless of duplicates in Grammy
unique_pairs = (
    matched_df[["wasabi_artist", "wasabi_title"]]
    .drop_duplicates()
    .reset_index(drop=True)
)
print(f"Unique (artist, title) pairs to query: {len(unique_pairs):,}")
already_cached = sum(
    1 for _, r in unique_pairs.iterrows()
    if _cache_key(r["wasabi_artist"], r["wasabi_title"]) in _cache
)
print(f"Already in cache: {already_cached:,}  |  Still to fetch: {len(unique_pairs) - already_cached:,}")

# Run queries
SAVE_EVERY = 100
new_queries = 0

for i, row in tqdm(unique_pairs.iterrows(), total=len(unique_pairs), desc="Last.fm"):
    result = query_lastfm(row["wasabi_artist"], row["wasabi_title"])
    new_queries += 1
    # Periodic cache flush
    if (i + 1) % SAVE_EVERY == 0:
        with open(CACHE_PATH, "w", encoding="utf-8") as fh:
            json.dump(_cache, fh)

# Final flush
with open(CACHE_PATH, "w", encoding="utf-8") as fh:
    json.dump(_cache, fh)

print(f"\n✅ Done. Cache size: {len(_cache):,} entries. Saved to {CACHE_PATH.name}")
```

### 3.7 Winner vs. Nominee Engagement by Category Group

For robust categories with Last.fm data: compare median listeners and playcount between winners and non-winning nominees.  
Uses Mann–Whitney U test (non-parametric, appropriate for skewed distributions).

```python
from scipy import stats

WINNER_MIN_N = 5  # require at least 5 winners with data to run the test

results_engagement = []

for grp in strat[strat["matched"] >= SPARSE_THRESHOLD].index:
    sub = matched_df[
        (matched_df["category_group"] == grp) &
        matched_df["lastfm_listeners"].notna()
    ]
    if sub.empty:
        continue

    winners_data   = sub[sub["winner"] == True]["lastfm_listeners"].dropna()
    nominees_data  = sub[sub["winner"] == False]["lastfm_listeners"].dropna()

    if len(winners_data) < WINNER_MIN_N or len(nominees_data) < WINNER_MIN_N:
        p_val = None
        direction = "—"
    else:
        stat, p_val = stats.mannwhitneyu(winners_data, nominees_data, alternative="two-sided")
        direction = "Winners > Nominees" if winners_data.median() > nominees_data.median() else "Nominees > Winners"

    results_engagement.append({
        "category_group"    : grp,
        "n_winners"         : len(winners_data),
        "n_nominees"        : len(nominees_data),
        "med_listeners_W"   : winners_data.median() if len(winners_data) > 0 else None,
        "med_listeners_N"   : nominees_data.median() if len(nominees_data) > 0 else None,
        "mann_whitney_p"    : round(p_val, 4) if p_val is not None else None,
        "sig_p05"           : "Yes" if p_val is not None and p_val < 0.05 else ("No" if p_val is not None else "—"),
        "direction"         : direction,
    })

eng_df = pd.DataFrame(results_engagement).sort_values("med_listeners_W", ascending=False, na_position="last")

print("Winner vs. Nominee engagement (Last.fm listeners) by category group:")
print(f"{'Group':<22} {'W_n':>5} {'N_n':>5}  {'Med W Listeners':>16}  {'Med N Listeners':>16}  {'p-val':>7}  {'Sig?':>5}  Direction")
print("-" * 110)
for _, r in eng_df.iterrows():
    mw = f"{r['med_listeners_W']:,.0f}" if pd.notna(r['med_listeners_W']) else "—"
    mn = f"{r['med_listeners_N']:,.0f}" if pd.notna(r['med_listeners_N']) else "—"
    pv = f"{r['mann_whitney_p']:.4f}" if pd.notna(r['mann_whitney_p']) else "—"
    print(f"{r['category_group']:<22} {r['n_winners']:>5} {r['n_nominees']:>5}  {mw:>16}  {mn:>16}  {pv:>7}  {r['sig_p05']:>5}  {r['direction']}")

print(f"\nNote: '—' means insufficient data (< {WINNER_MIN_N} winners with Last.fm coverage).")
```

### RQ1 Verification: Genre–Category Association

**Question:** What WASABI genres are most associated with Grammy-nominated artists, and how are these connected to the different categories?


```python
# ── RQ1: Genre–Category Association ──────────────────────────────────────────
# Use album_genre (82.8% coverage) as primary genre; fall back to genre column
df["genre_used"] = df["album_genre"].fillna(df["genre"])

rq1 = df.dropna(subset=["genre_used", "category_group"]).copy()
print(f"Rows with genre data: {len(rq1):,}  ({len(rq1)/len(df)*100:.1f}% of matched set)")

# Top category groups (top 15 by count)
top_cats = rq1["category_group"].value_counts().head(15).index.tolist()

# Top genres (top 12 by count)
top_genres = rq1["genre_used"].value_counts().head(12).index.tolist()

# Subset to top cats × top genres
rq1_sub = rq1[rq1["category_group"].isin(top_cats) & rq1["genre_used"].isin(top_genres)]

# Cross-tabulation: rows = category_group, cols = genre
cross = pd.crosstab(rq1_sub["category_group"], rq1_sub["genre_used"])

# ── Top 5 (genre, category) pairs by frequency ────────────────────────────────
pair_counts = (
    rq1_sub.groupby(["category_group", "genre_used"])
    .size()
    .reset_index(name="count")
    .sort_values("count", ascending=False)
)
print("\nTop 10 (category_group, genre) associations:")
for i, row in pair_counts.head(10).iterrows():
    print(f"  {i+1 if False else pair_counts.index.get_loc(i)+1}. ({row['genre_used']}, {row['category_group']}): {row['count']:,} occurrences")

# ── Chi-square test of independence ───────────────────────────────────────────
chi2, p_val, dof, expected = stats.chi2_contingency(cross)

print(f"\nChi-square test result:")
print(f"  χ² = {chi2:.2f}, df = {dof}, p-value = {p_val:.2e}")
if p_val < 0.05:
    print("  Interpretation: Genres and categories ARE significantly associated ✓")
    rq1_pass = True
else:
    print("  Interpretation: Genres and categories are NOT significantly associated ✗")
    rq1_pass = False

# ── Heatmap ────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 7))
sns.heatmap(
    cross,
    annot=True, fmt="d", cmap="YlOrRd",
    linewidths=0.3, ax=ax,
    cbar_kws={"label": "Nomination count"},
)
ax.set_title(f"RQ1: Grammy Category × Genre Co-occurrence\n(χ²={chi2:.1f}, p={p_val:.1e}  →  {'✓ Significant' if rq1_pass else '✗ Not significant'})",
             fontsize=12)
ax.set_xlabel("Genre (album_genre / genre)")
ax.set_ylabel("Grammy Category Group")
plt.xticks(rotation=40, ha="right", fontsize=9)
plt.yticks(rotation=0, fontsize=9)
plt.tight_layout()
plt.show()

print(f"\nRQ1 VERDICT: {'✅ PASS' if rq1_pass else '❌ FAIL'}")
```

### RQ2 Verification: Winner vs. Nominee Comparison

**Question:** For each Grammy category, how do genre distributions, production characteristics (BPM, loudness), and listener engagement differ between Grammy winners and non-winning nominees?


```python
# ── RQ2: Winner vs. Nominee Comparison — Production + Engagement + Genre ──────
SPARSE_THRESHOLD = 20
WINNER_MIN_N = 5

def safe_mw(a, b):
    a, b = a.dropna(), b.dropna()
    if len(a) < WINNER_MIN_N or len(b) < WINNER_MIN_N:
        return None
    _, p = stats.mannwhitneyu(a, b, alternative="two-sided")
    return p

# Category groups with enough matched rows
robust_groups = (
    df.groupby("category_group").size()
    .loc[lambda s: s >= SPARSE_THRESHOLD].index.tolist()
)

rq2_results = []
sig_groups = {"engagement": [], "bpm": [], "gain": [], "genre_diff": []}

for grp in robust_groups:
    sub = df[df["category_group"] == grp]
    w = sub[sub["winner"] == True]
    n = sub[sub["winner"] == False]

    # Engagement
    p_eng = safe_mw(w["lastfm_listeners"], n["lastfm_listeners"])
    med_eng_w = w["lastfm_listeners"].median()
    med_eng_n = n["lastfm_listeners"].median()

    # Production
    p_bpm  = safe_mw(w["bpm"],  n["bpm"])
    p_gain = safe_mw(w["gain"], n["gain"])
    med_bpm_w  = w["bpm"].median()
    med_bpm_n  = n["bpm"].median()
    med_gain_w = w["gain"].median()
    med_gain_n = n["gain"].median()

    # Genre distribution: top 3 genres for winners vs nominees
    def top3(series, col="genre_used"):
        vals = series[col].dropna().value_counts().head(3).index.tolist()
        return vals

    sub2 = sub.copy()
    sub2["genre_used"] = sub2["album_genre"].fillna(sub2["genre"])
    top_w = top3(w.assign(genre_used=w["album_genre"].fillna(w["genre"])))
    top_n = top3(n.assign(genre_used=n["album_genre"].fillna(n["genre"])))
    genre_overlap = len(set(top_w) & set(top_n))
    genre_diff = genre_overlap < 2  # fewer than 2 shared genres in top-3 = different

    rq2_results.append({
        "category_group" : grp,
        "n_winners"      : len(w),
        "n_nominees"     : len(n),
        "med_eng_W"      : med_eng_w,
        "med_eng_N"      : med_eng_n,
        "p_engagement"   : p_eng,
        "sig_eng"        : p_eng is not None and p_eng < 0.05,
        "med_bpm_W"      : med_bpm_w,
        "med_bpm_N"      : med_bpm_n,
        "p_bpm"          : p_bpm,
        "sig_bpm"        : p_bpm is not None and p_bpm < 0.05,
        "med_gain_W"     : med_gain_w,
        "med_gain_N"     : med_gain_n,
        "p_gain"         : p_gain,
        "sig_gain"       : p_gain is not None and p_gain < 0.05,
        "top_genres_W"   : top_w,
        "top_genres_N"   : top_n,
        "genre_diff"     : genre_diff,
    })

    if p_eng  is not None and p_eng  < 0.05: sig_groups["engagement"].append(grp)
    if p_bpm  is not None and p_bpm  < 0.05: sig_groups["bpm"].append(grp)
    if p_gain is not None and p_gain < 0.05: sig_groups["gain"].append(grp)
    if genre_diff: sig_groups["genre_diff"].append(grp)

rq2_df = pd.DataFrame(rq2_results).sort_values("n_winners", ascending=False)

# ── Print detailed report ─────────────────────────────────────────────────────
print("=" * 80)
print("RQ2 VERIFICATION: Winner vs. Nominee Comparison")
print("=" * 80)

for _, r in rq2_df.iterrows():
    print(f"\nCategory Group: {r['category_group']}  (n_winners={r['n_winners']}, n_nominees={r['n_nominees']})")

    # Engagement
    if pd.notna(r["p_engagement"]):
        sig = "✓ Significant" if r["sig_eng"] else "✗ Not significant"
        print(f"  Engagement (Last.fm listeners):")
        print(f"    Winners median:  {r['med_eng_W']:>10,.0f}")
        print(f"    Nominees median: {r['med_eng_N']:>10,.0f}")
        print(f"    p-value: {r['p_engagement']:.4f}  {sig}")
    else:
        print(f"  Engagement: insufficient data")

    # BPM
    bpm_w = f"{r['med_bpm_W']:.0f}" if pd.notna(r['med_bpm_W']) else "—"
    bpm_n = f"{r['med_bpm_N']:.0f}" if pd.notna(r['med_bpm_N']) else "—"
    if r["p_bpm"] is not None:
        sig = "✓ Significant" if r["sig_bpm"] else "✗ Not significant"
        print(f"  Production (BPM): Winners {bpm_w} vs Nominees {bpm_n}  p={r['p_bpm']:.4f}  {sig}")
    else:
        print(f"  Production (BPM): insufficient data")

    # Gain
    gain_w = f"{r['med_gain_W']:.1f}" if pd.notna(r['med_gain_W']) else "—"
    gain_n = f"{r['med_gain_N']:.1f}" if pd.notna(r['med_gain_N']) else "—"
    if r["p_gain"] is not None:
        sig = "✓ Significant" if r["sig_gain"] else "✗ Not significant"
        print(f"  Loudness (Gain):  Winners {gain_w} vs Nominees {gain_n}  p={r['p_gain']:.4f}  {sig}")
    else:
        print(f"  Loudness (Gain):  insufficient data")

    # Genres
    diff_note = "→ Genre distributions DIFFER" if r["genre_diff"] else "→ Genre distributions similar"
    print(f"  Top genres (Winners):  {r['top_genres_W']}")
    print(f"  Top genres (Nominees): {r['top_genres_N']}")
    print(f"  {diff_note}")

# ── Summary ───────────────────────────────────────────────────────────────────
n_rob = len(robust_groups)
print("\n" + "=" * 80)
print("RQ2 SUMMARY")
print("=" * 80)
print(f"Robust category groups: {n_rob}")
print(f"  Significant engagement difference: {len(sig_groups['engagement'])}/{n_rob}  {sig_groups['engagement']}")
print(f"  Significant BPM difference:        {len(sig_groups['bpm'])}/{n_rob}  {sig_groups['bpm']}")
print(f"  Significant Gain difference:       {len(sig_groups['gain'])}/{n_rob}  {sig_groups['gain']}")
print(f"  Different top genres:              {len(sig_groups['genre_diff'])}/{n_rob}  {sig_groups['genre_diff']}")

any_sig_per_group = sum(
    1 for r in rq2_results
    if r["sig_eng"] or r["sig_bpm"] or r["sig_gain"] or r["genre_diff"]
)
pct_sig = any_sig_per_group / n_rob * 100
rq2_pass = pct_sig >= 50
print(f"\nGroups with ≥1 significant difference: {any_sig_per_group}/{n_rob} ({pct_sig:.0f}%)")
print(f"RQ2 VERDICT: {'✅ PASS' if rq2_pass else '❌ FAIL'}  (threshold: ≥50% of groups show meaningful differences)")
```

### RQ3 Verification: Multi-Category Artists

**Question:** Which artists are connected to multiple Grammy categories, and how do the genres, BPM, loudness, and listener engagement of their nominated tracks differ across those categories?


```python
# ── RQ3: Multi-Category Artists ───────────────────────────────────────────────
# Normalise artist name for grouping (strip whitespace/case)
df["artist_norm_rq3"] = df["artist"].str.strip().str.lower()

artist_cats = (
    df.groupby("artist_norm_rq3")["category_group"]
    .nunique()
    .rename("n_categories")
)
artist_total = len(artist_cats)

multi_cat = artist_cats[artist_cats >= 2]
multi3_cat = artist_cats[artist_cats >= 3]

print("=" * 70)
print("RQ3: Multi-Category Artist Analysis")
print("=" * 70)
print(f"Total unique artists in dataset : {artist_total:,}")
print(f"Artists with ≥2 distinct categories: {len(multi_cat):,}  ({len(multi_cat)/artist_total*100:.1f}%)")
print(f"Artists with ≥3 distinct categories: {len(multi3_cat):,}  ({len(multi3_cat)/artist_total*100:.1f}%)")
print(f"Artists with exactly 2 categories  : {(len(multi_cat) - len(multi3_cat)):,}")

# ── Sample multi-category artists ─────────────────────────────────────────────
print("\nTop 15 multi-category artists (by number of distinct categories):")
top_multi = (
    multi_cat.sort_values(ascending=False)
    .head(15)
    .reset_index()
)
# Add nomination counts per artist
nom_counts = df.groupby("artist_norm_rq3").size().rename("total_nominations")
top_multi = top_multi.merge(nom_counts, on="artist_norm_rq3")

# Add the original artist name and category list
def get_cats(norm_name):
    return ", ".join(sorted(df[df["artist_norm_rq3"] == norm_name]["category_group"].unique()))

def get_display_name(norm_name):
    return df[df["artist_norm_rq3"] == norm_name]["artist"].iloc[0]

top_multi["display_name"] = top_multi["artist_norm_rq3"].map(get_display_name)
top_multi["categories"]   = top_multi["artist_norm_rq3"].map(get_cats)

for i, row in top_multi.iterrows():
    print(f"  {i+1}. {row['display_name']}:  {row['n_categories']} categories,  {row['total_nominations']} nominations")
    print(f"       Categories: {row['categories']}")

# ── Per-category nomination depth for multi-category artists ──────────────────
print("\nNomination depth for multi-category artists:")
multi_cat_names = multi_cat.index.tolist()
mc_df = df[df["artist_norm_rq3"].isin(multi_cat_names)].copy()

# For each (artist, category_group), count nominations
depth = (
    mc_df.groupby(["artist_norm_rq3", "category_group"])
    .size()
    .reset_index(name="n_nominations")
)
# Per artist: min nominations across categories they span
min_depth_per_artist = depth.groupby("artist_norm_rq3")["n_nominations"].min()

n_3plus  = (min_depth_per_artist >= 3).sum()
n_2plus  = (min_depth_per_artist >= 2).sum()
n_total  = len(min_depth_per_artist)

print(f"  Multi-category artists total          : {n_total:,}")
print(f"  With ≥3 nominations in both categories: {n_3plus:,}  ({n_3plus/n_total*100:.1f}%)")
print(f"  With ≥2 nominations in both categories: {n_2plus:,}  ({n_2plus/n_total*100:.1f}%)")
print(f"  With only 1 nomination in one category : {n_total - n_2plus:,}  ({(n_total-n_2plus)/n_total*100:.1f}%)")

# ── Distribution plot ──────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

artist_cats_plot = artist_cats[artist_cats >= 2]
axes[0].hist(artist_cats_plot, bins=range(2, artist_cats_plot.max()+2), color="steelblue", edgecolor="white", align="left")
axes[0].set_xlabel("Number of distinct Grammy categories")
axes[0].set_ylabel("Number of artists")
axes[0].set_title("Distribution: # Categories per Multi-Category Artist")
axes[0].set_xticks(range(2, int(artist_cats_plot.max())+1))

min_depth_plot = min_depth_per_artist.clip(upper=10)
axes[1].hist(min_depth_plot, bins=range(1, 12), color="darkorange", edgecolor="white", align="left")
axes[1].set_xlabel("Min nominations per category (clipped at 10)")
axes[1].set_ylabel("Number of artists")
axes[1].set_title("Nomination Depth: Multi-Category Artists")
axes[1].axvline(3, color="red", linestyle="--", label="≥3 threshold")
axes[1].legend()

plt.suptitle("RQ3: Multi-Category Artist Profile", fontsize=12, fontweight="bold")
plt.tight_layout()
plt.show()

# ── Feasibility verdict ────────────────────────────────────────────────────────
Y = len(multi_cat)
print("\n" + "=" * 70)
print("RQ3 FEASIBILITY ASSESSMENT")
print("=" * 70)
print(f"Total qualifying artists (≥2 categories): {Y}")
if Y >= 100:
    verdict = "FEASIBLE — statistical within-artist comparisons possible"
    rq3_pass = True
elif Y >= 50:
    verdict = "BORDERLINE — descriptive analysis only; no robust statistical testing"
    rq3_pass = True
else:
    verdict = "TOO SPARSE — rewrite RQ3 or merge with descriptive case studies"
    rq3_pass = False

print(f"VERDICT: {verdict}")
print(f"RQ3 VERDICT: {'✅ PASS' if rq3_pass else '❌ FAIL'}")
```
